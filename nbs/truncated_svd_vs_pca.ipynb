```python
from sklearn.decomposition import TruncatedSVD
from sklearn.decomposition import PCA
from sklearn.preprocessing import normalize

import pandas as pd
import scipy 
import numpy as np

def df(a): 
    return pd.DataFrame(np.round(a, 2))
```

```python
m, n = 7, 10
X = np.random.randn(m, n)
df(X)
```

## PCA

```python
pca = PCA(n_components=10)
df(pca.fit_transform(X))
```

```python
pca = PCA(n_components=9)
df(pca.fit_transform(X))
```

```python
pca = PCA(n_components=6)
df(pca.fit_transform(X))
```

```python
pca.explained_variance_ratio_
```

```python
pca = PCA(n_components=4)
df(pca.fit_transform(X))
```

```python
pca.singular_values_
```

```python
pca.explained_variance_ratio_
```

```python
df(pca.components_)
```

## TruncatedSVD

```python
trun_svd = TruncatedSVD(n_components=9)
df(trun_svd.fit_transform(X))
```

```python
trun_svd = TruncatedSVD(n_components=5)
df(trun_svd.fit_transform(X))
```

```python
trun_svd = TruncatedSVD(n_components=4)
df(trun_svd.fit_transform(X))
```

```python
trun_svd.singular_values_
```

```python
trun_svd.explained_variance_ratio_
```

```python
df(trun_svd.components_)
```

```python
pca = PCA(n_components=4)
trun_svd = TruncatedSVD(n_components=4)
df(trun_svd.fit_transform(X) - pca.fit_transform(X))
```
