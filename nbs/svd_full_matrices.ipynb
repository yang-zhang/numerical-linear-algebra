```python
import pandas as pd
import scipy 
import numpy as np
from scipy.linalg import svd

def df(a): 
    return pd.DataFrame(np.round(a, 2))
```

## `m < n`

```python
m, n = 4, 10
X = np.random.randn(m, n)
df(X)
```

### `full_matrices=True`

```python
U, s, Vh = svd(X, full_matrices=True)
X.shape, U.shape, s.shape, Vh.shape
```

```python
S = np.zeros((m, n))
S[:m, :m] = np.diag(s)
X.shape, U.shape, S.shape, Vh.shape
```

```python
df(U)
```

```python
df(S)
```

```python
np.allclose(U @ S @ Vh, X)
```

```python
df(U @ S)
```

```python
df(U * s)
```

```python
df(U @ np.diag(s))
```

```python
df(Vh)
```

```python
df(Vh[:m])
```

```python
df(U * s @ Vh[:m])
```

```python
np.allclose(U * s @ Vh[:m], X)
```

```python
Vh_ = np.copy(Vh)
Vh_[m:] = 0
df(Vh_)
```

```python
np.allclose(U @ S @ Vh_, X)
```

### `full_matrices=False`

```python
U, s, Vh = svd(X, full_matrices=False)
X.shape, U.shape, s.shape, Vh.shape
```

```python
np.allclose(U * s @ Vh, X)
```

```python
df(U * s)
```

```python
S = np.diag(s)
X.shape, U.shape, S.shape, Vh.shape
```

```python
df(U @ S)
```

```python
df(Vh)
```

```python
df(U @ S @ Vh)
```

```python
df(U * s @ Vh)
```

```python
np.allclose(U @ S @ Vh, X), np.allclose(U * s @ Vh, X)
```

## `m > n`

```python
m, n = 10, 4
X = np.random.randn(m, n)
df(X)
```

### `full_matrices=True`

```python
U, s, Vh = svd(X, full_matrices=True)

X.shape, U.shape, s.shape, Vh.shape
```

```python
S = np.zeros((m, n))
S[:n, :n] = np.diag(s)
X.shape, U.shape, S.shape, Vh.shape
```

```python
df(S)
```

```python
np.allclose(U @ S @ Vh, X)
```

```python
df(S @ Vh)
```

```python
df(np.diag(s) @ Vh)
```

```python
df(s * Vh)
```

Notice!

```python
np.allclose(df(s * Vh), df(np.diag(s) @ Vh))
```

```python
df(U[:, :n])
```

```python
np.allclose(U[:, :n] @ (np.diag(s) @ Vh), X)
```

Notice!

```python
np.allclose(U[:, :n] @ (s * Vh), X)
```

```python
df(U[:, :n] * s)
```

```python
df(U[:, :n] @ np.diag(s))
```

```python
np.allclose(U[:, :n] * s @ Vh, X)
```

```python
U_ = np.copy(U)
U_[:, n:] = 0
df(U_)
```

```python
np.allclose(U_ @ S @ Vh, X)
```

### `full_matrices=False`

```python
U, s, Vh = svd(X, full_matrices=False)

X.shape, U.shape, s.shape, Vh.shape
```

```python
np.allclose(U * s @ Vh, X)
```
