```python
import pandas as pd
import scipy 
import numpy as np
from scipy.linalg import svd

def df(a): 
    return pd.DataFrame(np.round(a, 2))

m, n = 5, 10
X = np.random.randn(m, n)
df(X)
```

### `full_matrices=True`

```python
U, s, Vh = svd(X, full_matrices=True)
S = np.zeros((m, n))
S[:m, :m] = np.diag(s)
np.allclose(U @ S @ Vh, X)
```

```python
for top_m in range(m, 0, -1):
    S_top_m = np.zeros_like(S)
    S_top_m[:top_m] = S[:top_m]
    print(S_top_m)
    print('\n')
    print(((U @ S_top_m @ Vh - X)**2).sum())
    print('\n')
```

```python
r = 3
df(U[:, :r] @ S[:r, :r] @ Vh[:r])
```

### `full_matrices=False`

```python
U, s, Vh = svd(X, full_matrices=False)

np.allclose(U * s @ Vh, X)
```

```python
for top_m in range(m, 0, -1):
    s_top_m = np.zeros_like(s)
    s_top_m[:top_m] = s[:top_m]
    print(s_top_m)
    print(((U * s_top_m @ Vh - X)**2).sum())
```

```python
k = 3
s_k = np.zeros_like(s)
s_k[:k] = s[:k]
df(U * s_k @ Vh)
```
